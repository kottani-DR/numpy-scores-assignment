```python
import pandas as pd

# Direct download link
url = "https://drive.google.com/uc?id=1Mlu9r3afSGz5_02Y3wDJoQLpXszQXaIU"

# Load dataset
df = pd.read_csv(url, index_col="order_id")

# First 5 rows
print("First 5 rows:")
display(df.head())

# Last 5 rows
print("Last 5 rows:")
display(df.tail())
```

```python
# Shape of dataset
print("Dataset Shape:", df.shape)

# Info about dataset
df.info()

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Statistical summary
summary = df.describe()
print(summary)

# Median unit_price
median_price = df["unit_price"].median()
print("\nMedian unit_price:", median_price)
```
